```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Leer CSV y convertir columna Fecha a datetime
df = pd.read_csv('base_datos.csv')
df['Fecha'] = pd.to_datetime(df['Fecha'], dayfirst=True, errors='coerce')
df = df.dropna(subset=['Fecha'])

# Agregar por intervalos de 2 meses
counts = df.set_index('Fecha').resample('2MS').size()

# Graficar
fig = plt.figure(figsize=(10,5))
plt.bar(counts.index.strftime('%Y-%m'), counts.values, color='C0')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Intervalo de 2 meses (fecha de inicio)')
plt.ylabel('Cantidad')
plt.title('Cantidad de reportes cada 2 meses')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()

# Guardar imagen ANTES de plt.show()
output_folder = "../frontend/img"
output_image = os.path.join(output_folder, "grafica_reportes.png")

# Guardar correctamente la figura
fig.savefig(output_image, dpi=300)

# Mostrar en notebook
plt.show()

print("Imagen generada en:", output_image)

```
